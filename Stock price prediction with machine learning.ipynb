```python
import yfinance as yf
```

```python
## LOADING IN THE DATA ##
#create a ticker object
sp = yf.Ticker("^GSPC")
```

```python
#lets query the historical prices of the object
sp = sp.history(period='max')
```

```python
sp
```

```python
sp.index
```

```python
## CLEANING AND EDA DATA ##
```

```python
sp.plot.line(y='Close', use_index=True)
```

```python
#lets remove the columns we dont need
del sp['Dividends']
del sp['Stock Splits']
```

```python
sp
```

```python
## Set up our target ##
#target is what the model will be predicting
#The idea is not to predict the exact price but rather if it will go up or down in price
#we create 2 more cols one for comparing the closing price of each day and one is bool wheter it go up or down

sp['tmr'] = sp['Close'].shift(-1)
sp['target'] = (sp['tmr'] > sp['Close']).astype(int)
sp
```

```python
# we also dw data from too long ago, as it might cause inaccuracies 
# lets remove all data from 1990 and older

sp = sp.loc['2000-01-01':].copy()
sp
```

```python
## LETS TRAIN OUR FIRST MODEL ##
# lets use random forest, this type of model is less prone to overfitting
from sklearn.ensemble import RandomForestClassifier

#initialise our model
model = RandomForestClassifier(n_estimators = 200, min_samples_split = 100, random_state=1)

#split our data
#note we cannot use crossvalidation on a time series data, or it will be using future data to predict past values
#we dont go tradional way of splitting, 8-2 but we jsut take last 100 days worth of data as test
train = sp.iloc[:-100]
test = sp.iloc[-100:]

#lets define which features are predictors
predictors = ["Close", "Volume", "Open", "High", "Low"]
# we are using the predictors col to predict the target
model.fit(train[predictors], train['target'])
```

```python
## LETS MEASURE ACCURACY OF OUR MODEL ##

from sklearn.metrics import precision_score
preds = model.predict(test[predictors])

# convert from numpy array to pandas series so its easier to work with
import pandas as pd
preds = pd.Series(preds, index=test.index)

precision_score(test['target'],preds)
```
